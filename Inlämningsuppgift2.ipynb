## 3. Vad var invånarantalet i varje region i mitten på år 2015? Använd Pandas för att visa.

```python
assult.groupby("Region_2013_label")[["Population_mid_point_2015"]].sum()\
.rename(columns={"Population_mid_point_2015": "Population"})
```

## 4. Hur många brott begicks i varje region år 2015? Ni hittar brotten i kolumnen ”Victimisations_calendar_year_2015”.

```python
assult.groupby("Region_2013_label")[["Region_2013_label","Victimisations_calendar_year_2015"]].sum()
```

## 5. Visa de 10 områden med de högsta antalet överfall. Områden är ”Area_unit_2013_label”.

```python
assultArea = assult.nlargest(10, "Victimisations_calendar_year_2015")\
[["Area_unit_2013_label", "Population_mid_point_2015", "Victimisations_calendar_year_2015"]]
assultArea
```

## 6. Lägg till en kolumn i dataframen som visar % antal överfall i området jämfört med invånarantal i området. Varje rad är ett område

```python
assultArea["%_Övefall_Invånare"]=(assultArea["Victimisations_calendar_year_2015"] /assultArea["Population_mid_point_2015"])\
.mul(100).round(2).astype(str) + "%"
assultArea
```

## 7. Skapa en ny dataframe som innehåller antalet områden (Area_unit_2013_label) som finns av de olika typerna av urban area (Urban_area_2013_label).

```python
urban_area = assult.groupby("Urban_area_2013_label").count()[["Area_unit_2013_label"]].head(20)
urban_area
```

## 8. Ta reda på hur många områden (varje rad är ett område) hade population men inga överfall. Jämfört totala populationen i dessa med totala populationen i deras region(Region_2013_label).

```python
#assult.sort_values(by=["Victimisations_calendar_year_2015" , "Population_mid_point_2015"],ascending=[True, False])
#assult[["Region_2013_label", "Victimisations_calendar_year_2015", "Population_mid_point_2015"]]\
#.sort_values("Victimisations_calendar_year_2015")

assult.loc[(assult["Victimisations_calendar_year_2015"] == 0) & (assult["Population_mid_point_2015"] >0)]\
[["Region_2013_label", "Victimisations_calendar_year_2015", "Population_mid_point_2015"]]\
.sort_values("Population_mid_point_2015")
```

## 9. Räkna per kolumn hur många rader det är som inte innehåller någon data

```python
assult[assult.isna().any(axis=1)]
```

## 10. Skapa en ny dataframe med 10 rader. Den innehåller de 10 rader med lägsta antalet överfall per invånare (Rate_per_10000_population), men mer än 5000 invånare. 

```python
lowest_assult = (assult["Population_mid_point_2015"] >5000 ) & (assult["Rate_per_10000_population"])
assult.loc[lowest_assult, ["Population_mid_point_2015" , "Rate_per_10000_population"]]\
.sort_values("Rate_per_10000_population").head(10)
```

## 11. Skapa en ny dataframe ”regions” som innehåller sammansatt data per region (Region_2013_label). Lägg till en ny kolumn som innehåller ”Rate_per_10000_population” fast för hela regionen

```python
regions = assult.groupby("Region_2013_label").sum()[["Victimisations_calendar_year_2015","Population_mid_point_2015"]]
regions["Rate_per_10000_population_region"]\
=(regions["Victimisations_calendar_year_2015"] / regions["Population_mid_point_2015"])\
.mul(10000).round(0).astype(str)
regions
```

## 12. Lägg till en ny kolumn i dataframen med regionerna som visar i % hur många av alla överfall i landet som sker i varje region

```python
regions["%_Övefall_pr_regionen"]=(regions["Victimisations_calendar_year_2015"] / regions.sum()["Victimisations_calendar_year_2015"])\
.mul(100).round(2).astype(str) + "%"
regions
```

## Data cleaning

## % av saknad data. Kollar hur många procent av data som saknas. Tycker det ger en bättre visualisering hur man kan analysera rätt data.

```python
for columns in edible_oils:
    procent = edible_oils[columns].isnull().mean()
    print(f"{columns}: {round(procent*100,2)}%")
```

## Ändar "Month" till datetime

```python
edible_oils["Month"]=pd.to_datetime(edible_oils["Month"])
```

## Ändra NaN värdet till 0
#### Sunflower Oil, Rapeseed Oil, Palm Kernel Oil saknar en del info om pris och change i %. Från augusti -92 till december -02.
### Istället för att sätta NaN värden till 0 så väljer jag att ta bort alla rader som har NaN värden.

```python
edible_oils.dropna(inplace=True)
```

```python
#edible_oils = edible_oils[edible_oils["Sunflower Oil Price"].notna()]
```

```python
#edible_oils["Palm Kernel Oil Price"].fillna(0, inplace=True), edible_oils["Palm Kernel Oil Price Change"].fillna(0, inplace=True)
```

## Ändrar Dtype till float på kolumner som jag vill ha

```python
olive = ["Olive Oil Price"]
for col in olive:
    edible_oils[col] = edible_oils[col].str.replace(",","").astype(float)
```

```python
sunflower = ["Sunflower Oil Price"]
for col in sunflower:
    edible_oils[col] = edible_oils[col].str.replace(",","").astype(float)
```

```python
rapeseed = ["Rapeseed Oil Price"]
for col in rapeseed:
    edible_oils[col] = edible_oils[col].str.replace(",","").astype(float)
```

```python
palm = ["Palm Oil Price"]
for col in palm:
    edible_oils[col] = edible_oils[col].str.replace(",","").astype(float)
```

## Priser på matolja

```python
oils=edible_oils.groupby("Month")[["Olive Oil Price","Sunflower Oil Price", "Rapeseed Oil Price", "Palm Oil Price"]].sum()
oils
```
